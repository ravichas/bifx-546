# Default Dictionary

A default Dictionary is like a regular dictionary, except that when you try to look up a key it doesnt contain, it first adds a value for it using a zero-argument function you provided when you created it. Note that you need to `import defaultdict` to use it

```
d = defaultdict(factory_function)
factory_functions: `int`, `float`, `list`, `set`, `dict`

```

````python
from collections import defaultdict

d = defaultdict(int) # int() -> 0, so default value is 0
print(d["a"]) # key "a" doesnt exist yet, so it returns 0
d["a"] += 5
print(d["a"]) # now "a" is 5
````

````python
document = ["data", "science", "from", "scratch"]

word_counts = defaultdict(int)          # int() produces 0
for word in document:
    word_counts[word] += 1

print(word_counts)
````

# Why defaultdict?

defaultdict(list) means: a dictionary where every new (missing) key automatically starts with an empty list [].

So you can append to keys without checking if the key already exists.

Normal dict (fails if key missing)

```
d = {}
d[0].append(1)   # KeyError, because d[0] doesn't exist yet
```

But the following code will work

```
from collections import defaultdict

d = defaultdict(list)
d[0].append(1)   # d[0] was automatically created as []
d[0].append(2)

print(d[0])      # [1, 2]
```

Why is it useful for us?

No need to pre-create friendships = {0: [], 1: [], ...} â€” it builds itself as you go.

````python
dd_list = defaultdict(list)             # list() produces an empty list
print(dd_list)
dd_list[2].append(1)                    # now dd_list contains {2: [1]}
print(dd_list)
````

````python
dd_dict = defaultdict(dict)             # dict() produces an empty dict
dd_dict["Joel"]["City"] = "Seattle"     # {"Joel" : {"City": Seattle"}}
print(dd_dict)

dd_pair = defaultdict(lambda: [0, 0])
dd_pair[2][1] = 1                       # now dd_pair contains {2: [0, 1]}
print(dd_pair)
````

# How `and` works
```
A and B
```

Python works this way:
* Evaluate `A`
* if `A` is *falsey* (`""`, `0`, `None`, `[]`, etc), return **A** immediately (dont evaluate **B**)
* if `A` is truthy, evaluate and return **B*


```
"" and "x"  # "" (left if falsey, so return left)
"hello" and "x". # "x"(left is truthy, so return right)
0 and 5 # 0
3 and 5 # 5
None and "x" # None
"yes" and "no" # "no"
```

## Object Oriented Programming

* Class contains zero or more member functions
* By convention each takes a first parameter `self` that refers to a particular instance
* Each class normally has a constructor `__init__` (method is often called "dunder" methods (double-UNDERscore), it takes whatever parameters you need to construct an instance of your class and does what ever setup you need:

```
def __init__(self, count=0):
    self.count = count
```





````python
def smallest_item(xs):
    assert xs, "empty list has no smallest item"
    return min(xs)

class CountingClicker:
    """A class can/should have a docstring, just like a function"""

    def __init__(self, count = 0):
        self.count = count

    def __repr__(self):
        return f"CountingClicker(count={self.count})"

    def click(self, num_times = 1):
        """Click the clicker some number of times."""
        self.count += num_times

    def read(self):
        return self.count

    def reset(self):
        self.count = 0

````

````python
clicker = CountingClicker()
assert clicker.read() == 0, "clicker should start with count 0"
clicker.click()
print(clicker.count)
clicker.click()
print(clicker.count)
assert clicker.read() == 2, "after two clicks, clicker should have count 2"
clicker.reset()
assert clicker.read() == 0, "after reset, clicker should be back to 0"
````

# Subclass that Inherits from the Main Class

````python
# A subclass inherits all the behavior of its parent class.
class NoResetClicker(CountingClicker):
    # This class has all the same methods as CountingClicker

    # Except that it has a reset method that does nothing.
    def reset(self):
        pass
````

# Iterables and Generators

A generator is a  specalized kind of generator that produces values one at a time, only when needed, instead of creating all in the memory

A generator function is any function that contains the keyword `yield`


````python
def generate_range(n):
    i = 0
    while i < n:
        yield i   # every call to yield produces a value of the generator
        i += 1

for i in generate_range(10):
    print(f"i: {i}")
````

Let us run this with some more information

List: High peak memory (it stores all `n` elements)
Generator: Much lower peak memory (only holds one item ata time plus overhead)

````python
import time
import tracemalloc

n = 10000000
print("===List===")
start = time.time()
lst = list(range(n))
end = time.time()
print("List creation time:", end-start, "seconds")
````

````python
start = time.time()

for _ in generate_range(n):
  pass

end = time.time()
print("Generator creation time:", end-start, "seconds")
````

````python
print("====List===")
tracemalloc.start()
lst = list(range(n))
current, peak = tracemalloc.get_traced_memory()
print(f"Current:{current/1_000_000:.2f}MB; Pak: {peak/1_000_000:.2f}MB")
tracemalloc.stop()
````

````python
print("====Generator===")
tracemalloc.start()
gen = generate_range(n)
for x in gen:
  pass
current, peak = tracemalloc.get_traced_memory()
print(f"Current:{current/1_000_000:.2f}MB; Pak: {peak/1_000_000:.2f}MB")
tracemalloc.stop()
````

In Python lazy means: the values are computed only when they are needed, not in advance



````python
def natural_numbers():
  """ return 1, 2, 3, .... """
  n = 1
  while True:
    yield n
    n += 1
````

````python
# None of these computations *does* anything until we iterate
data = natural_numbers()
evens = (x for x in data if x % 2 == 0)
even_squares = (x ** 2 for x in evens)
even_squares_ending_in_six = (x for x in even_squares if x % 10 == 6)
# and so on


assert next(even_squares_ending_in_six) == 16
assert next(even_squares_ending_in_six) == 36
assert next(even_squares_ending_in_six) == 196
````

# A better example  
### Let us put generators that we created above to action

````python
for x in even_squares_ending_in_six:
  print(x)
  if x > 1000:
    break
````

Step by step:
* for loop asks `even_squares_ending_in_six` for the next value
* that generator asks `even_squares` for the next value
* `even_squares` asks `evens` for the next value
* `evens` asks `data` (natural numbers for the next number
* Each step does the needed computaiton for **one** value, then returns it.

# Use `enumerate` instead of `for` or `range(len(list))`

````python
names = ["Alice", "Bob", "Charlie", "Debbie"]

# not Pythonic
for i in range(len(names)):
    print(f"name {i} is {names[i]}")
print()

# also not Pythonic
i = 0
for name in names:
    print(f"name {i} is {names[i]}")
    i += 1
print()

# Pythonic
for i, name in enumerate(names):
    print(f"name {i} is {name}")
````

# Regular Expressions

* Effecient way of searching text
* Useful but complicated syntax


Note
* `re.matxh` checks the beginning of a string matches a regular expression
* `re.search` checks whether any part of  a string matches a reg expression



````python
import re

re_examples = [                        # all of these are true, because
    not re.match("a", "cat"),              #  'cat' doesn't start with 'a'
    re.search("a", "cat"),                 #  'cat' has an 'a' in it
    not re.search("c", "dog"),             #  'dog' doesn't have a 'c' in it
    3 == len(re.split("[ab]", "carbs")),   #  split on a or b to ['c','r','s']
    "R-D-" == re.sub("[0-9]", "-", "R2D2") #  replace digits with dashes
    ]

assert all(re_examples), "all the regex examples should be True"
````

# `arg` and `kwargs`

How to create higher-order functions that takes some function as inputs and returns a new function

````python
def doubler(f):
    # Here we define a new function that keeps a reference to f
    def g(x):
        return 2 * f(x)

    # And return that new function.
    return g

def f1(x):
    return x + 1

g = doubler(f1)

assert g(3) == 8,  "(3 + 1) * 2 should equal 8"
assert g(-1) == 0, "(-1 + 1) * 2 should equal 0"
````

````python
# rewrote becos the function name and variable name, g confusing

def doubler(f):
    def doubled_function(x):
        return 2 * f(x)
    return doubled_function

def f1(x):
    return x + 1

g = doubler(f1)   # g now refers to doubled_function

assert g(3) == 8,  "(3 + 1) * 2 should equal 8"
assert g(-1) == 0, "(-1 + 1) * 2 should equal 0"
````

````python
def f2(x, y):
    return x + y

g = doubler(f2)
````

````python
try:
    g(1, 2)
except TypeError:
    print("as defined, g only takes one argument")
````

# the above example takes only one arg

How to expand this to multiple arguments

````python
def magic(*args, **kwargs):
    print("unnamed args:", args)
    print("keyword args:", kwargs)

magic(1, 2, key="word", key2="word2")

# prints
#  unnamed args: (1, 2)
#  keyword args: {'key': 'word', 'key2': 'word2'}
````

````python
def other_way_magic(x, y, z):
    return x + y + z

x_y_list = [1, 2]
z_dict = {"z": 3}
assert other_way_magic(*x_y_list, **z_dict) == 6, "1 + 2 + 3 should be 6"

def doubler_correct(f):
    """works no matter what kind of inputs f expects"""
    def g(*args, **kwargs):
        """whatever arguments g is supplied, pass them through to f"""
        return 2 * f(*args, **kwargs)
    return g

g = doubler_correct(f2)
assert g(1, 2) == 6, "doubler should work now"
````

# Type Annotations

Python is a dynamically typed language. It doesnt care about the type of objects we use, as long as we use them in valid ways.

````python
def add(a, b):
    return a + b

assert add(10, 5) == 15,                  "+ is valid for numbers"
assert add([1, 2], [3]) == [1, 2, 3],     "+ is valid for lists"
assert add("hi ", "there") == "hi there", "+ is valid for strings"

try:
    add(10, "five")
except TypeError:
    print("cannot add an int to a string")
````

In a statically typed language our functions and objects have to be specified this way. Modern Python allow this.

````python
def add(a: int, b: int) -> int:
    return a + b

add(10, 5)           # you'd like this to be OK
add("hi ", "there")  # you'd like this to be not OK
````

### Since Python has first-class functions, we need to type to represent those as well. Here is an example,

````python
from typing import Callable

# The type hint says that repeater is a function that takes
# two arguments, a string and an int, and returns a string.
def twice(repeater: Callable[[str, int], str], s: str) -> str:
    return repeater(s, 2)

def comma_repeater(s: str, n: int) -> str:
    n_copies = [s for _ in range(n)]
    return ', '.join(n_copies)

assert twice(comma_repeater, "type hints") == "type hints, type hints"
````
