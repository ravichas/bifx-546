There are 2 ways to create plots using Matplotlib

* pyplot: Designed to mimic Matplot. This approach is for creating fast and simple plots.
* OO Style: Designed for creating flexible, complex plots

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Simple commands (Matplotlib tracks the 'state' in the background)
plt.plot(x, y, color='blue', marker='o')
plt.title("Pyplot Style")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

## Object-oriented Style

Commonly used for creating complex layouts


```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Explicitly create the objects
# ax is a child of the fig, modifying ax directly updates the content
# within the fig.
fig, ax = plt.subplots()

# Call methods on the 'ax' object
ax.plot(x, y, color='red', marker='s')
ax.set_title("Object-Oriented Style")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```

 If you ever run plt.subplots(2, 2), ax becomes a list (array) of four different plot areas, while fig remains the single window holding them all.


```python
ax.get_children() # get all Artists objects living under ax
```

```python
# creating a grid of plots
import matplotlib.pyplot as plt

plt.style.use('ggplot')

# Create a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Access specific plots using [row, column] indices
axs[0, 0].plot([1, 2], [1, 2], 'r') # Top-left
axs[0, 1].scatter([1, 2], [2, 1])   # Top-right
axs[1, 0].bar([1, 2], [3, 4])      # Bottom-left
axs[1, 1].hist([1, 1, 2, 3, 3, 3]) # Bottom-right

# Batch styling using a loop
for ax in axs.flat:
    ax.set(xlabel='X Label', ylabel='Y Label')

plt.tight_layout()

plt.show()
```
