## 🎯 The Core Problem

In machine learning, we are almost always trying to **minimize something** — usually a **loss function** that measures how wrong our model is.

The question is:

> *How do we find the model parameters that make the loss as small as possible?*

Sometimes calculus gives us an exact answer — set the derivative to zero and solve. This is called a **closed-form solution**.

But for most real ML problems, a closed-form solution is either:
- mathematically impossible, or
- computationally impractical (too many parameters, too complex a function)

We need a **numerical approach**. That approach is **gradient descent**.

## 🏔️ The Key Idea: Walk Downhill

Gradient descent is essentially **navigating a landscape blindfolded**.

You can't see the whole terrain, but you *can* feel the slope under your feet.

- The **gradient** tells you which direction is *uphill* the fastest.
- To go downhill, you take a small step in the **opposite direction** of the gradient.
- Repeat this enough times and you'll eventually reach a valley — a **minimum**.

The update rule is:

$$\theta \leftarrow \theta - \eta \, \nabla L(\theta)$$

where:
- $\theta$ = model parameters
- $\eta$ = learning rate (step size)
- $\nabla L(\theta)$ = gradient of the loss with respect to the parameters

---

> 💡 **Analogy from Joel Grus:** Think of it like standing on a foggy mountain. You can't see the bottom, but you can feel which way the ground slopes. You take a small step downhill. Then another. Eventually you reach the valley.

## 🧱 Three Ingredients You Need

Every gradient descent problem requires three things:

### 1. A Function to Minimize — the Loss
For example, **Mean Squared Error (MSE)** for linear regression:

$$L(m, b) = \frac{1}{n} \sum_{i=1}^{n} (m x_i + b - y_i)^2$$

### 2. The Gradient of That Function
How the loss changes as each parameter changes — computed via partial derivatives:

$$\frac{\partial L}{\partial m} = \frac{2}{n} \sum_{i=1}^{n} (m x_i + b - y_i)\, x_i$$

$$\frac{\partial L}{\partial b} = \frac{2}{n} \sum_{i=1}^{n} (m x_i + b - y_i)$$

### 3. A Step Size — the Learning Rate $\eta$
Controls how large each update step is. Choosing this well is critical (see below).

## 🍦 Three Flavors of Gradient Descent

Joel Grus introduces three variants, each making a different tradeoff between **accuracy of the gradient estimate** and **speed of updates**:

| Variant | Data Used Per Update | Updates Per Epoch | Character |
|---|---|---|---|
| **Batch GD** | All data | 1 | Stable but slow |
| **Mini-batch GD** | Small subset | Many | Balanced — used in practice |
| **Stochastic GD (SGD)** | 1 example | $n$ | Noisy but very fast |

---

Explanations:

"Data Used Per Update" — how many training examples are used to compute the gradient before taking one step.

"Updates Per Epoch" — an epoch is one full pass through your dataset. This column tells you how many parameter updates happen in that one pass.

"Character" — the practical personality of that variant.

---

### Row by Row

**Batch GD**

Looks at all 100 data points, averages their gradients, then takes one step.
Like polling the entire country before making a decision — very reliable, but slow.

**Mini-batch GD**

Splits the data into small chunks (e.g., 20 at a time), updates after each chunk.
If you have 100 data points and batch size = 20, you get 5 updates per epoch.
**The sweet spot — used in virtually all modern deep learning.**

**Stochastic GD (SGD)**

Looks at just one example at a time, updates immediately.
If you have n = 100 data points, you get 100 updates per epoch.
Very fast, but the gradient estimate is noisy — it zigzags toward the minimum rather than walking smoothly.

### The Core Tradeoff

- **More data per step** → smoother, more accurate gradient → fewer steps needed, but each step is expensive
- **Less data per step** → noisier gradient → more steps needed, but each step is cheap and updates happen frequently

In practice, **mini-batch GD** is the standard choice in modern deep learning.

Epoch 1:

```
  Point 1  → compute gradient → update θ   (1st update)
  Point 2  → compute gradient → update θ   (2nd update)
  Point 3  → compute gradient → update θ   (3rd update)
  ...
  Point 100 → compute gradient → update θ  (100th update)
```

Epoch 2: shuffle, repeat...

## ⚙️ The Learning Rate Problem

Choosing the step size $\eta$ is more **art than science** — as Grus notes in the chapter.

| Learning Rate | Effect |
|---|---|
| Too **small** | Descent is correct but takes forever |
| Too **large** | Overshoots the minimum; loss may *increase* or diverge |
| Just right | Converges smoothly to the minimum |

Common strategies:
- **Fixed step size** — simplest; works well if chosen carefully
- **Shrinking step size** — reduce $\eta$ over time as you get closer to the minimum
- **Line search** — find the optimal step size at each iteration (expensive)

Throughout the book, Grus mostly uses a **fixed learning rate** and experiments to find a good value.

## 🔄 The Gradient Descent Loop

Regardless of the variant, every gradient descent algorithm follows the same loop:

```
1. Initialize parameters θ randomly
2. Repeat for many epochs:
      a. Compute the gradient ∇L(θ) on some data
      b. Update: θ ← θ - η × ∇L(θ)
3. Return θ when loss is small enough (or after fixed epochs)
```

## 🌍 Why This Chapter Matters

Gradient descent is the **engine behind nearly all of modern machine learning**.

| Model | What GD Minimizes |
|---|---|
| Linear Regression | Mean Squared Error |
| Logistic Regression | Log Loss (Cross-Entropy) |
| Neural Networks | Any differentiable loss |
| Dose-Response Fitting | Nonlinear squared error |

The specific loss functions and gradients change across models — but **the core loop is always the same**:

> Compute gradient → Step downhill → Repeat
