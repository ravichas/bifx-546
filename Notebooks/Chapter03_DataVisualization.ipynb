# Discussion in Class

1Ô∏è‚É£ Choose the Right Chart

*	What type of data do you have?
  -	Categories
  -	Distributions
  -	Time series
  -	Proportions
*	Who is the audience?
*	Follow best practices:
  *	Simple charts (line charts, bar plots with a shared axis) are generally easier to interpret
  *	Challenging or often misused charts include pie charts
  *	Avoid inappropriate chart choices (e.g., using scatter plots when bar charts are more suitable)

2Ô∏è‚É£ Simplify the Chart

*	Add clear annotations, labels, and titles
*	Split complex visuals into multiple charts
*	Use gridding or faceting when comparing groups
*	Use animation sparingly and only when it adds meaning
*	Use arrows or callouts only when they guide interpretation

3Ô∏è‚É£ Text and Labels

*	Font size, style, and color matter
*	Title considerations:
  *	Is a bold title necessary?
  *	Is the title too long or too short?
  *	Is the title grammatically correct and free of redundancy?
*	Are units included on all axes?
*	Are axis labels readable without excessive rotation?

4Ô∏è‚É£ Focus the Viewer‚Äôs Attention

*	Remove nonessential elements
*	Use gray or muted colors for secondary information
*	Check whether legends are crowded and could be replaced with direct labels
*	Avoid distracting background colors
*	Ask whether grid lines are necessary
*	If boxes or borders are present, do they add value?

````python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Bucket grades by decile, but put 100 in with the 90s
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

print(histogram) # Counter({80: 4, 90: 3, 70: 3, 0: 2, 60: 1})

# Counter will produce dict with k, v
# plot.bar([list-upper], [list-lower], )
plt.bar([x + 5 for x in histogram.keys()],  # Shift bars right by 5
        histogram.values(),                 # Give each bar its correct height
        10,                                 # Give each bar a width of 10
        edgecolor=(0, 0, 0))                # Black edges for each bar

plt.axis([-5, 105, 0, 5])                  # x-axis from -5 to 105,
                                           # y-axis from 0 to 5

plt.xticks([10 * i for i in range(11)])    # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
# plt.show()


plt.savefig('im/viz_grades.png')
# plt.gca().clear()
````

# Key points to note
* `plt.axis`: indicates the x-axis range (-5 to 105, in this case; note the little space on the Left/Right)
* `plt.xticks` puts x-axis labels
* Note how we start from x-axis at 0


# Using SNS

````python
# Ensure all deciles from 0 to 90 are present (even if frequency = 0)
all_deciles = list(range(0, 100, 10))
data = {
    "Decile": all_deciles,
    "Frequency": [histogram.get(d, 0) for d in all_deciles]
}

df = pd.DataFrame(data)
# Make a string label so seaborn treats it nicely as categorical
df["DecileLabel"] = df["Decile"].astype(str)
df
````

````python
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=df, x="DecileLabel", y="Frequency", color="skyblue", edgecolor="black")

ax.set_xlabel("Decile")
ax.set_ylabel("# of Students")
ax.set_title("Distribution of Exam 1 Grades")

plt.show()
````

````python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)

# misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
# plt.show()


plt.savefig('im/viz_misleading_y_axis.png')
# plt.gca().clear()

````

````python
variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# We can make multiple calls to plt.plot
# to show multiple series on the same chart
plt.plot(xs, variance,     'g-',  label='variance')    # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2')      # red dot-dashed line
plt.plot(xs, total_error,  'b:',  label='total error') # blue dotted line

# Because we've assigned labels to each series,
# we can get a legend for free (loc=9 means "top center")
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The Bias-Variance Tradeoff")
# plt.show()


plt.savefig('im/viz_line_chart.png')
# plt.gca().clear()
````

## Scatter Plots

````python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
        xy=(friend_count, minute_count), # Put the label with its point
        xytext=(5, -5),                  # but slightly offset
        textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
# plt.show()


plt.savefig('im/viz_scatterplot.png')
# plt.gca().clear()
````
